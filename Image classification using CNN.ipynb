```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import glob
import shutil
from PIL import Image
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dense,MaxPooling2D, Flatten,Dropout
```

```python
data_dir="E:\HK\ML\Image classification"
classes=['BMW','Mercedes','Audi','Rangerover','Koenigsegg']
for category in classes:
    path=os.path.join(data_dir,category)
   
    image=glob.glob(path+ '/*.jfif')
    
    train_len=int(round(len(image)*0.7))
    train, test=image[:train_len],image[train_len:]
    for t in train:
        if not os.path.exists(os.path.join(data_dir,'train',category)):
           os.makedirs(os.path.join(data_dir,'train',category))
        shutil.move(t,os.path.join(data_dir,'train',category))
    for s in test:
        if not os.path.exists(os.path.join(data_dir,'test',category)):
               os.makedirs(os.path.join(data_dir,'test',category))
        shutil.move(s,os.path.join(data_dir,'test',category))
    
```

```python
train_dir=os.path.join(data_dir,'train')
test_dir=os.path.join(data_dir,'test')
```

```python
for c in classes:
    for file in glob.glob(os.path.join(train_dir,c)+'/*.jfif'):
        im=Image.open(file)
        rgb_im = im.convert('RGB')
        rgb_im.save(file.replace("jfif","jpg"),quality=95)
```

```python
for c in classes:
    for file in glob.glob(os.path.join(test_dir,c)+'/*.jfif'):
        im=Image.open(file)
        rgb_im = im.convert('RGB')
        rgb_im.save(file.replace("jfif","jpg"),quality=95)
```

```python
batch_size=150
img_shape=150
img_gen=ImageDataGenerator(rescale=1./255,horizontal_flip=True)
train_image=img_gen.flow_from_directory(batch_size=batch_size,directory=train_dir,shuffle=True,target_size=(img_shape,img_shape))
img_gen=ImageDataGenerator(rescale=1./255, shear_range=0.3,rotation_range=45,horizontal_flip=True)
train_image=img_gen.flow_from_directory(batch_size=batch_size,directory=train_dir,shuffle=True,target_size=(img_shape,img_shape),class_mode='sparse')
```

```python
testgen=ImageDataGenerator(rescale=1./255)
test_image=testgen.flow_from_directory(batch_size=batch_size,directory=test_dir,shuffle=True,target_size=(img_shape,img_shape),class_mode="sparse")
```

```python
test_image.n
```

```python
model= Sequential()
model.add(Conv2D(16,3,padding='same',activation='relu',input_shape=(img_shape,img_shape,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32,3,padding='same',activation='relu',input_shape=(img_shape,img_shape,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,3,padding='same',activation='relu',input_shape=(img_shape,img_shape,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(5,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
epochs=120
his=model.fit_generator(train_image, steps_per_epoch=int(np.ceil((train_image.n)/(float(batch_size)))),epochs=epochs,validation_data=test_image, validation_steps=int(np.ceil(test_image.n/(float(batch_size)))))
```
